```python
# Main imports
import numpy as np
import pandas as pd

# Import matplotlib and such
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
plt.ion()

from helper_functions import \
    thermodynamic_model, \
    plot_manifold_model, \
    plot_manifold_measurements, \
    get_measurement_subset_df
    
```

```python
# Set global styles
textwidth=5.5
halfcolwidth=3
fontsize=8
titlesize=10
panel_label_size=12
legend_fontsize=7
markersize=5
mpl.rcParams['font.size'] = fontsize
mpl.rcParams['text.usetex'] = False

# Get default colors
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

# Used to create color gradient across data points of different P values
color_strong = np.array([.11, .69, .93])
color_weak = np.array([1.00, .94, .21])

# Conversion parameter
kbt_to_kcal = 1/1.62

# From previous analyses
F_50 = 23.89230  # FINAL 18.11.08
log_tsat_fixed =  2.722414412  # FINAL 18.11.08
t_sat_fixed = np.exp(log_tsat_fixed)
```

```python
# Load clonal measurements 
data_df = pd.read_excel('../data/results.xlsx', sheet_name='measurements_summary').set_index('name')
data_df.head()
```

```python
c41_resamp_df = pd.read_excel('../data/results.xlsx', sheet_name='c41_beta_resamp').set_index('run')
c41_resamp_df.head()
```

```python
### Figure 9: class II regulation

fig, _ = plt.subplots(figsize=[textwidth,3.75])
gs = gridspec.GridSpec(1, 2, top=.70, bottom=.12, left=.12, right=.98, hspace=0, wspace=.5)

### Panel B

# P values for continuous curve
Ps = np.logspace(-7,1,100)

# Parameters for best c41 curve
t_bg_best = np.exp(c41_resamp_df.loc['fit','log_tbg'])
alphap_best = np.exp(c41_resamp_df.loc['fit','log_alphap'])
betap_best = np.exp(c41_resamp_df.loc['fit','log_betap'])
alpha_best = alphap_best*(1+1/F_50)-1/F_50
beta_best = betap_best*(1+1/(alpha_best*F_50)) - 1/(alpha_best*F_50)

# Parameters for sampled c41 curve
t_bg_sampled = np.exp(c41_resamp_df.loc[:,'log_tbg'].values[1:])
alphap_sampled = np.exp(c41_resamp_df.loc[:,'log_alphap'].values[1:])
betap_sampled = np.exp(c41_resamp_df.loc[:,'log_betap'].values[1:])
alpha_sampled = alphap_sampled*(1+1/F_50)-1/F_50
beta_sampled = betap_sampled*(1+1/(alpha_sampled*F_50)) - 1/(alpha_sampled*F_50)

ax = plt.subplot(gs[0,0])

# Plot styles
samples_labels_colors = [
    ('gal.', '-41.5 bp', 'deepskyblue'),
]


# Plot sampled models
for (t_bg, alpha, beta) in zip( t_bg_sampled, alpha_sampled, beta_sampled):
    plot_manifold_model(ax=ax, 
                    t_sat=t_sat_fixed,
                    t_bg=t_bg,
                    alpha=alpha,
                    beta=beta,
                    F=F_50,
                    color='silver',
                    opacity=.1
                    )
print('%d bootstrap samples'%len(alpha_sampled))


# Plot fits
plot_manifold_model(ax=ax, 
                    t_sat=t_sat_fixed,
                    t_bg=t_bg_best,
                    alpha=alpha_best,
                    beta=beta_best,
                    F=F_50,
                    color='black',
                    opacity=1,
                    )


# Plot measurements
ax.axhline(t_sat_fixed,linestyle='--',color='gray')
ax.axvline(t_sat_fixed,linestyle='--',color='gray')

n1, o1 = plot_manifold_measurements(ax=ax, 
                           df=data_df, 
                           samples_labels_colors=samples_labels_colors[:1], 
                           markersize=markersize, 
                           fontsize=fontsize,
                           show_errorbars=False)

ax.set_aspect('equal')



# xs = thermodynamic_model(t_sat=t_sat, P=Ps, F=0, alpha=alpha_best, beta=1, t_bg=t_bg)
# ys = thermodynamic_model(t_sat=t_sat, P=Ps, F=F_50, alpha=alpha_best, beta=1, t_bg=t_bg)
# #ax.plot(xs,ys,'-',color=colors[0],zorder=-100, alpha=.5)

### Panel B

ax = plt.subplot(gs[0,1])

# Plot styles
samples_labels_colors = [
    ('c40.10L', '-40.5 bp', 'yellowgreen'),
]

# Plot measurements
# ax.axhline(t_sat,linestyle='--',color='gray')
# ax.axvline(t_sat,linestyle='--',color='gray')

n3, o3 = plot_manifold_measurements(ax=ax, 
                           df=data_df, 
                           samples_labels_colors=samples_labels_colors, 
                           markersize=markersize, 
                           fontsize=fontsize,
                           show_errorbars=True,
                           show_dots=True)

ax.set_aspect('equal')


### Add panel labels
plt.figtext(.01,.98,'(A)', fontsize=panel_label_size, horizontalalignment='left', verticalalignment='top')
plt.figtext(.01,.70,'(B)', fontsize=panel_label_size, horizontalalignment='left', verticalalignment='top')
plt.figtext(.52,.98,'(C)', fontsize=panel_label_size, horizontalalignment='left', verticalalignment='top')
plt.figtext(.52,.70,'(D)', fontsize=panel_label_size, horizontalalignment='left', verticalalignment='top')

fig.savefig('../figures/plt_classII.pdf', transparent=True)
```
