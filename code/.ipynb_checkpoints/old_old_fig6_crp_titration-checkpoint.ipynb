```python
# Main imports
import numpy as np
import pandas as pd

# Import matplotlib and such
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline

from scipy import stats
from sklearn.linear_model import LinearRegression

from helper_functions import \
    thermodynamic_model, \
    plot_manifold_model, \
    plot_manifold_measurements, \
    get_clonal_measurements_df, \
    get_resampled_params_df, \
    get_distance_params_df,\
    get_measurement_subset_df
```

```python
# Set global styles
textwidth=5.5
halfcolwidth=3
fontsize=8
titlesize=10
panel_label_size=12
legend_fontsize=7
markersize=5
mpl.rcParams['font.size'] = fontsize
mpl.rcParams['text.usetex'] = False

# Get default colors
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

# Used to create color gradient across data points of different P values
color_strong = np.array([.11, .69, .93])
color_weak = np.array([1.00, .94, .21])

# Conversion parameter
kbt_to_kcal = 1/1.62

# Set t_sat, t_bg, and alpha
log_t_sat = 2.723033948
log_t_bg_occ = -6.060653019
log_t_bg_c71 = -5.606744091
log_alpha_c71 = 4.624254371
```

```python
# Load data
data_df = pd.read_excel('../data/results.xlsx', sheet_name='cAMP_summary')

# Create crp_spacing column
data_df['crp_spacing'] = np.nan
for i in data_df.index:
    name = data_df.loc[i,'name']
    if name[:3] == 'c71':
        data_df.loc[i,'crp_spacing'] = -71.5
    elif name[:3] == 'oc0':
        data_df.loc[i,'crp_spacing'] = +0.5        
        
# Remove dlog columns
columns = data_df.columns
columns = [c for c in columns if not 'dlog_t' in c]
# columns = [c for c in columns if not '125.0' in c]
# columns = [c for c in columns if not '25.0' in c]
# columns = [c for c in columns if not '5.0' in c]
data_df = data_df[columns]

# Show data_df
data_df.head()
```

```python
# Create dict of colors
concs = [250,50,10,2.5]
full = np.array([.80, .48, .93])
none = np.array([1, 1, 1])
colors_dict = {
    250: .66*full,
    50: full,
    10: .66*full + .33*none,
    2.5: .33*full + .66*none
}

# Samples labels colors
oc0_slc = [('%.1f_oc0.*'%conc, '%.1f $\\mu$M'%conc, colors_dict[conc]) for conc in concs]
c71_slc = [('%.1f_c71.*'%conc, '%.1f $\\mu$M'%conc, colors_dict[conc]) for conc in concs]

# Create new df that is compatible with plotting tools
new_df = pd.DataFrame(columns=['name','log_t+','log_t-','outlier'])
for i, row in data_df.iterrows():
    for conc in concs:
        new_row_dict = {
            'name':'%.1f_%s'%(conc,row['name']),
            'log_t+':row['log_t_%.1f'%conc],
            'log_t-':row['log_t_0.0'],
            'outlier':False,
            'dlog_t+':0,
            'dlog_t-':0
        }
        new_df = new_df.append(new_row_dict, ignore_index=True)

new_df = new_df.dropna()
new_df = new_df.set_index('name')
new_df
```

```python
# Load fitting results
results_df = pd.DataFrame(columns=['conc','crp_spacing','log_F_50','log_F_16','log_F_84'])
for conc in [250,125,50,25,10,5,2.5]:
             
    # Occlusion data
    sheet_name = 'occlusion_resamp_%.1fuM'%conc
    tmp_df = pd.read_excel('../data/results.xlsx', sheet_name=sheet_name)
    log_Fs = tmp_df['log_F'].values[1:]
    row = {
        'conc':conc,
        'crp_spacing':0.5,
        'log_F_50':np.percentile(log_Fs,50),
        'log_F_16':np.percentile(log_Fs,16),
        'log_F_84':np.percentile(log_Fs,84),
    }
    results_df = results_df.append(row, ignore_index=True)
             
    # Recruitment data
    sheet_name = 'c71r18_resamp_%.1f'%conc
    tmp_df = pd.read_excel('../data/results.xlsx', sheet_name=sheet_name)
    log_Fs = tmp_df['log_F'].values[1:]
    row = {
        'conc':conc,
        'crp_spacing':-71.5,
        'log_F_50':np.percentile(log_Fs,50),
        'log_F_16':np.percentile(log_Fs,16),
        'log_F_84':np.percentile(log_Fs,84),
    }
    results_df = results_df.append(row, ignore_index=True)
        
grouped_results_df = results_df.groupby(['conc','crp_spacing']).first()
grouped_results_df
```

```python
### Figure 2: Occlusion measurements ##
fig, axs = plt.subplots(figsize=[textwidth,6])

gs = gridspec.GridSpec(1, 2, top=.80, bottom=.45, left=.12, right=.98, wspace=.2)


# Set lims
lim = [1E-3,3E1]
ticks = np.logspace(-3,2)

# Subplot A: occlusion
ax = plt.subplot(gs[0,0])

# Plot data points
num_points, num_outliers = plot_manifold_measurements(ax=ax, 
                               df=new_df, 
                               samples_labels_colors=oc0_slc, 
                               markersize=markersize, 
                               fontsize=fontsize,
                               lim=lim,
                               legend_loc='upper left')

# Plot fit models
for conc in concs:
    log_F = grouped_results_df.loc[(conc,0.5),:]['log_F_50']
    plot_manifold_model(ax, 
                       t_sat = np.exp(log_t_sat),
                       t_bg = np.exp(log_t_bg_occ),
                       F = np.exp(log_F),
                       alpha = 0,
                       beta = 1,
                       color=colors_dict[conc])
ax.set_title('CRP at +0.5 bp')


# Subplot B: c71 
ax = plt.subplot(gs[0,1])

num_points, num_outliers = plot_manifold_measurements(ax=ax, 
                               df=new_df, 
                               samples_labels_colors=c71_slc, 
                               markersize=markersize, 
                               fontsize=fontsize,
                               lim=lim,
                               legend_loc='lower right')

df = data_df.copy()
df = df[df['crp_spacing']==-71.5]
xs = df['log_t_0.0'].values
for conc in concs:
    color = colors_dict[conc]
    col = 'log_t_%.1f'%conc
    ys = df[col].values
    
    indices = np.isfinite(ys)
    ax.plot(xs,ys,'o',markersize=markersize,label='%.1f $\mu$M'%conc, color=color)

# Plot fit models
for conc in concs:
    log_F = grouped_results_df.loc[(conc,-71.5),:]['log_F_50']
    plot_manifold_model(ax, 
                       t_sat = np.exp(log_t_sat),
                       t_bg = np.exp(log_t_bg_c71),
                       F = np.exp(log_F),
                       alpha = np.exp(log_alpha_c71),
                       beta = 1,
                       color=colors_dict[conc])
ax.set_yticklabels([])
ax.set_ylabel('')
ax.set_title('CRP at -71.5 bp')

# Plot C
gs = gridspec.GridSpec(1, 1, top=.32, bottom=.08, left=.33, right=.75, wspace=0)
ax = plt.subplot(gs[0,0])

# # Grab Fs
xlim = np.array([1E0,1E3])
ylim = np.array([1E-2,1E2])
spacings = [0.5,-71.5]
labels = ['CRP at +0.5 bp','CRP at -71.5 bp']
colors_to_use = [colors[1],colors[2]]
concs_lists = [(250,125,50,25,10,5.0,2.5), (250,125,50,25,10,5,2.5)]

slopes = []
intercepts = []
for spacing, color, label, concs_to_keep in zip(spacings, colors_to_use, labels, concs_lists):
    Fs_50 = np.exp(np.array([grouped_results_df.loc[(conc,spacing),:]['log_F_50'] for conc in concs_to_keep]))
    Fs_16 = np.exp(np.array([grouped_results_df.loc[(conc,spacing),:]['log_F_16'] for conc in concs_to_keep]))
    Fs_84 = np.exp(np.array([grouped_results_df.loc[(conc,spacing),:]['log_F_84'] for conc in concs_to_keep]))

    # Fit line
    reg = LinearRegression()
    log_xs = np.log(concs_to_keep) 
    log_ys = np.log(Fs_50)
    
    # Plot
    for (conc,F_16,F_84) in zip(concs_to_keep, Fs_16, Fs_84):
        ax.loglog([conc, conc],[F_16, F_84], '-', color=color)
    ax.loglog(concs_to_keep,Fs_50,'s',color=color, label=label)

    slope, intercept, r_value, p_value, std_err = stats.linregress(log_xs,log_ys)
    slopes.append(slope)
    intercepts.append(intercept)
    line_log_xs = np.log(xlim)
    line_log_ys = intercept + slope*line_log_xs
    
    ax.loglog(np.exp(line_log_xs), np.exp(line_log_ys), '-', color=color)
    print('%s: power= %.2f +- %.2f'%(label,slope,1.96*std_err))
    
# Print estimated fractional difference
conc = 25
F_act = np.exp(intercepts[1] + slopes[1]*np.log(conc))
F_rep = np.exp(intercepts[0] + slopes[0]*np.log(conc))
f = F_rep / F_act
print('F_rep / F_act at %.1f uM: %.1f'%(conc,f))
    
ax.set_ylim(ylim)
ax.set_xlim(xlim)
ax.set_xlabel('[cAMP] ($\mu$M)')
ax.set_ylabel('$F$ (unitless)')
ax.legend(loc='upper left', fontsize=fontsize)

# Label panels
plt.figtext(.01,.98,'(A)', fontsize=panel_label_size, horizontalalignment='left', 
            verticalalignment='top')
#plt.figtext(.53,.98,'(C)', fontsize=panel_label_size, horizontalalignment='left', 
#            verticalalignment='top')
plt.figtext(.01,.83,'(B)', fontsize=panel_label_size, horizontalalignment='left', 
            verticalalignment='top')
plt.figtext(.53,.83,'(C)', fontsize=panel_label_size, horizontalalignment='left', 
            verticalalignment='top')
plt.figtext(.20,.34,'(D)', fontsize=panel_label_size, horizontalalignment='left', 
            verticalalignment='top')

#plt.tight_layout()
fig.savefig('../figures/6_crp_titration.pdf', transparent=True)
```
