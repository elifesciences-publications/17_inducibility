```python
# Main imports
import numpy as np
import pandas as pd

# Import matplotlib and such
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
plt.ion()

from helper_functions import \
    thermodynamic_model, \
    plot_manifold_model, \
    plot_manifold_measurements, \
    get_measurement_subset_df
    
```

```python
# Set global styles
textwidth=5.5
halfcolwidth=3
fontsize=8
titlesize=10
panel_label_size=12
legend_fontsize=7
markersize=5
mpl.rcParams['font.size'] = fontsize
mpl.rcParams['text.usetex'] = False

# Get default colors
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

# Used to create color gradient across data points of different P values
color_strong = np.array([.11, .69, .93])
color_weak = np.array([1.00, .94, .21])

# Conversion parameter
kbt_to_kcal = 1/1.62

# From occlusion results
F_50 = 23.89230  # FINAL 18.11.08
```

```python
# Load clonal measurements 
data_df = pd.read_excel('../data/results.xlsx', sheet_name='measurements_summary').set_index('name')
data_df.head()
```

```python
# Load c61 meassurements
c61_model_df = pd.read_excel('../data/results.xlsx', sheet_name='c61_resamp').set_index('run')
c61_model_df.head()
```

```python
# Load spacing results
distance_df = pd.read_excel('../data/results.xlsx', sheet_name='conjoined_resamp').set_index('run')
distance_df.head()
```

```python
# Tally up the number of promoters used for each inference
locations = [c.split('_')[0] for c in distance_df.columns if '_log_P' in c]
len(locations)

df = data_df.copy()
df.set_index('location', inplace=True)
df = df.loc[locations,:]
spacings = list(set(df['spacing'].values))
spacings.sort(reverse=True)

spacing_to_n_dict = {}
for spacing in spacings:
    n = sum(df['spacing'].values == spacing)
    print('d = %.1f, n = %d'%(spacing, n))
    spacing_to_n_dict[spacing] = n
```

```python
# Prepare data frame to hold recruitment table
columns = [
    'position', 
    'num_inliers', 
    'num_outliers',
    'dG', 
    'alpha', 
    'gaston',
    'ushida',
]
gaston_dict = {
    -60.5:'3.85',
    -61.5:'9.05',
    -62.5:'4.22',
    -66.5:'0.78',
    -71.5:'2.50',
    -72.5:'3.49',
    -76.5:'0.54'
}
ushida_dict = {
    -61.5:'20.6',
    -66.5:'0.84',
    -71.5:'16.4',
    -82.5:'6.99'
}
table_df = pd.DataFrame(columns=columns)
table_df.head()
```

```python
### Figure 4: Recruitment data ###
fig, _ = plt.subplots(figsize=[textwidth,6])

lim = [5E-4,2E2]

### Panel B

gs = gridspec.GridSpec(1, 1, top=.80, bottom=.45, left=.15, right=.52)
ax = plt.subplot(gs[0,0])

color10 = colors[1]
color35 = np.array([.99, .50, .16])*.7

# Plot styles
samples_labels_colors = [
    ('c61r18.35L', '-35 series', color35),
    ('c61r18.10L', '-10 series', color10),
]

# Extract best fit model
t_sat_best = np.exp(c61_model_df.loc['fit','log_tsat'])
t_bg_best = np.exp(c61_model_df.loc['fit','log_tbg'])
alphap_best = np.exp(c61_model_df.loc['fit','log_alphap'])

# Extract sampled fits
t_sat_sampled = np.exp(c61_model_df['log_tsat'].values[1:])
t_bg_sampled = np.exp(c61_model_df['log_tbg'].values[1:])
alphap_sampled = np.exp(c61_model_df['log_alphap'].values[1:])

# Compute corresponding alpha values; don't propagate errors though. 
alpha_best = alphap_best + (alphap_best-1)/F_50
alpha_sampled = alphap_sampled + (alphap_sampled-1)/F_50
alpha_50 = np.percentile(alpha_sampled,50)
alpha_84 = np.percentile(alpha_sampled,84)
alpha_16 = np.percentile(alpha_sampled,16)

# Plot sampled models
for (t_sat, t_bg, alpha) in zip(t_sat_sampled, t_bg_sampled, alpha_sampled):
    plot_manifold_model(ax=ax, 
                    t_sat=t_sat,
                    t_bg=t_bg,
                    alpha=alpha,
                    F = F_50,
                    color='silver',
                    opacity=.1,
                    )
print('%d bootstrap samples'%len(alpha_sampled))

# Plot best fit model
plot_manifold_model(ax=ax, 
                    t_sat=t_sat_best,
                    t_bg=t_bg_best,
                    alpha=alpha_best,
                    F = F_50,
                    color='black',
                    #label='fit',
                    )
# Plot measurements
num_points, num_outliers = plot_manifold_measurements(ax=ax, 
                               df=data_df, 
                               samples_labels_colors=samples_labels_colors, 
                               markersize=markersize, 
                               fontsize=fontsize,
                               lim=lim,
                               ticks=[1E-3,1E-2,1E-1,1E0,1E1,1E2])

# Display Delta G
#plt.title('$\Delta G_\\alpha = %0.2f \pm %0.2f$ kcal/mol'%(c61_DeltaG,c61_dDeltaG))
#ax.text(x=1E0, y=3E-2, s='$\\alpha = 601^{+76}_{-86}$')
ax.set_xlim(lim)
ax.set_ylim(lim)
ax.set_aspect('equal')

# Compute energy for c61
kbt_to_kcal = 1/1.62

c61_DeltaG = -kbt_to_kcal*np.log(alpha_50)
c61_dDeltaG = kbt_to_kcal*.5*(np.log(alpha_84)-np.log(alpha_16))
print('-61.5: DeltaG = %0.2f +/- %0.2f'%(c61_DeltaG, c61_dDeltaG))

dalpha_plus = alpha_84 - alpha_50
dalpha_minus = alpha_50 - alpha_16

text_1 = '$\\Delta G_\\alpha = %0.2f \\pm %0.2f$ kcal/mol\n'%(c61_DeltaG, c61_dDeltaG)
text_2 = '$\\alpha = %d^{+%d}_{-%d}$\n '%(alpha_50, dalpha_plus, dalpha_minus)
#print(text_1)
print(text_2)
ax.text(x=lim[0]*2, y=lim[1]/2, s=text_1, fontsize=fontsize,
        horizontalalignment='left', 
        verticalalignment='top')
ax.text(x=3E-1, y=1E-1, s=text_2, fontsize=fontsize,
        horizontalalignment='left', 
        verticalalignment='top')
#ax.set_title('CRP at -61.5 bp')

### Panel C
gs = gridspec.GridSpec(6, 2, top=.95, bottom=.02, left=.62, right=.98, wspace=.0, hspace=.0)

samples_dists = [
    ('c60',60.5), 
    ('c61',61.5),
    ('c62',62.5), 
    ('c63',63.5), 
    ('c64',64.5),
    ('c65',65.5),
    ('c66',66.5),
    ('c71',71.5), 
    ('c72',72.5),
    ('c76',76.5),
    ('c81',81.5),
    ('c82',82.5)]

# Get default color list
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']


t_sat = np.exp(distance_df.loc['fit','log_tsat'])
t_sats = np.exp(distance_df.loc[:,'log_tsat'].values[1:])
t_sat_50 = np.percentile(t_sats,50)
t_sat_84 = np.percentile(t_sats,84)
t_sat_16 = np.percentile(t_sats,16)

print('t_sat = %.1f^{+%.1f}_{-%.1f}'%(t_sat_50,t_sat_84-t_sat_50,t_sat_50-t_sat_16))

slc = []
for n, pair in enumerate(samples_dists):
    row = int(n/2)
    col = n%2
    
    ax = plt.subplot(gs[row,col])
    
    sample = pair[0]
    dist = pair[1]
    
    color = colors[n%len(colors)]
    #x.set_aspect(1)
    samples_labels_colors = [(sample,None,color)]
    slc.extend(samples_labels_colors)
    
    t_bg = np.exp(distance_df.loc['fit','log_tbg_%s'%sample])
    alpha = np.exp(distance_df.loc['fit','log_alphap_%s'%sample])
    
    # Plot model
    plot_manifold_model(ax, t_sat=t_sat, t_bg=t_bg, alpha=alpha, F=F_50, color='black')
    
    # Plot measurements, 
    num_points, num_outliers = plot_manifold_measurements(ax,data_df,samples_labels_colors, 
                                                           markersize=3, xlabel='', ylabel='', 
                                                           fontsize=9, show_legend=False, lim=lim)
    ax.set_xticks([])
    ax.set_yticks([])
    
    ## Fill in table
    # Load results from distance_df
    from_row = 'fit'
    from_col = 'log_alphap_%s'%sample
    log_alphaps = distance_df.loc[:,from_col].values[1:]
    
    # alphap = (1+F*alpha)/(1+F)
    # (1+F)*alphap = 1+F*alpha
    # (1+1/F)*alphap = 1/F + alpha
    # alpha = alphap*(1+1/F) - 1/F
    
    # Compute alpha from alphap
    alphaps = np.exp(log_alphaps)
    alphas = alphaps*(1+1/F_50) - 1/F_50
    
    # Compute alphas and dGs
    dGs = -kbt_to_kcal*np.log(alphas)
    alpha_50 = np.percentile(alphas,50) 
    alpha_84 = np.percentile(alphas,84)
    alpha_16 = np.percentile(alphas,16)
    dalpha_minus = alpha_50-alpha_16
    dalpha_plus = alpha_84-alpha_50
    
    dG_50 = np.percentile(dGs,50)
    dG_16 = np.percentile(dGs,16)
    dG_84 = np.percentile(dGs,84)
    ddG = 0.5*(dG_84-dG_16)
    
    # Write strings
    dG_str = r'$%.2f\pm%.2f$'%(dG_50,ddG)
    if alpha > 100:
        alpha_str = '$%d^{+%d}_{-%d}$'%(alpha_50,dalpha_plus,dalpha_minus)
    elif alpha > 10:
        alpha_str = '$%.1f^{+%.1f}_{-%.1f}$'%(alpha_50,dalpha_plus,dalpha_minus)
    else:
        alpha_str = '$%.2f^{+%.2f}_{-%.2f}$'%(alpha_50,dalpha_plus,dalpha_minus)
    
    # Fill out table
    table_df.loc[n,'position'] = '%.1f'%(-dist)
    table_df.loc[n,'num_inliers'] = '%d'%spacing_to_n_dict[-dist]
    table_df.loc[n,'num_outliers'] = '%d'%num_outliers
    table_df.loc[n,'alpha'] = alpha_str
    table_df.loc[n,'dG'] = dG_str    
    table_df.loc[n, 'gaston'] = gaston_dict[-dist] if -dist in gaston_dict.keys() else '-'
    table_df.loc[n, 'ushida'] = ushida_dict[-dist] if -dist in ushida_dict.keys() else '-'
    
    # Set title
    ax.text(x=1E-1, y=3E-4, s='-%.1f bp\n'%dist)
    #ax.set_title('-%.1f bp'%dist)
    
print('slc: %s'%repr(slc))    
    
### Panel D

gs = gridspec.GridSpec(1, 1, top=.35, bottom=.08, left=.15, right=.52)
ax = plt.subplot(gs[0,0])

# Draw stems 
for n, pair in enumerate(samples_dists):
    
    sample = pair[0]
    dist = -pair[1]
    
    alphas = np.exp(distance_df.loc[:,'log_alphap_%s'%sample].values)[1:]
    alpha_50 = np.percentile(alphas,50)
    alpha_16 = np.percentile(alphas,16)
    alpha_84 = np.percentile(alphas,84)
    
    # Plot point
    ax.errorbar([dist],[alpha_50],
            yerr=[[alpha_84-alpha_50], [alpha_50-alpha_16]], 
            linestyle='none', marker='o', 
            markersize=3, capsize=markersize, color='k')
    
    # Plot stem
    ax.plot([dist,dist], [.1,alpha_50], '-', linewidth=1, zorder=-100)

# Show minor tick marks on x
from matplotlib.ticker import AutoMinorLocator
minorLocator = AutoMinorLocator()
ax.xaxis.set_minor_locator(minorLocator)

xlim = np.array([-84,-59])
alpha_lims = np.array([3E-1,1E3])
ax.set_yscale("log", nonposy='clip')
ax.set_ylim(alpha_lims)
ax.axhline(1, linestyle='--', color='k', linewidth=1)
ax.set_ylabel('cooperativity  $\\alpha$')
ax.set_xlim(xlim)
ax.set_xlabel('CRP site center')
    
energy_lims = np.log(alpha_lims)/1.62
ax2 = ax.twinx()
ax2.set_ylim(energy_lims)
ax2.set_ylabel('$\\Delta G_\\alpha$   [kcal/mol]')
ax2.set_yticks([0,1,2,3,4])
ax2.set_yticklabels([0,-1,-2,-3,-4])

    
# Tight layout
panel_label_size=12
plt.figtext(.01,.98,'(A)', fontsize=panel_label_size, horizontalalignment='left', verticalalignment='top')
plt.figtext(.01,.82,'(B)', fontsize=panel_label_size, horizontalalignment='left', verticalalignment='top')
plt.figtext(.55,.98,'(C)', fontsize=panel_label_size, horizontalalignment='left', verticalalignment='top')
plt.figtext(.01,.39,'(D)', fontsize=panel_label_size, horizontalalignment='left', verticalalignment='top')
    
fig.savefig('../figures/plt_recruitment_results.pdf', transparent=True)
```
