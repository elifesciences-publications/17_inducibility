```python
import pylab as py
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import scipy.optimize as opt
import sys
from random import randint
from random import uniform
import itertools

%matplotlib inline
```

```python
cAMP_loaded_data=pd.read_csv('../data/old_cAMP/cAMP_data.txt', sep='\t', header=0)
cAMP_summary=pd.read_csv('../data/old_cAMP/cAMP_fit.txt', sep='\t', header=0)
```

```python
median_vals={'c61r18':{},'c71r18':{},'occlusion':{}}
outliers=[]

for row,sample in enumerate(cAMP_loaded_data['name']):
    loc=cAMP_loaded_data['loc'][row]
    
    if not cAMP_loaded_data['inlier'][row]:
        outliers.append(loc)
    
    if 'c61' in sample:
        median_vals['c61r18'][loc]={'0.0':cAMP_loaded_data['bas_0.0'][row]}
        for conc in ['250.0','125.0','50.0','25.0','10.0','5.0','2.5']:
            median_vals['c61r18'][loc][conc]=cAMP_loaded_data['ind_'+conc][row]
        
    if 'c71' in sample:
        median_vals['c71r18'][loc]={'0.0':cAMP_loaded_data['bas_0.0'][row]}
        for conc in ['250.0','125.0','50.0','25.0','10.0','5.0','2.5']:
            median_vals['c71r18'][loc][conc]=cAMP_loaded_data['ind_'+conc][row]
        
    if 'oc' in sample:
        median_vals['occlusion'][loc]={'0.0':cAMP_loaded_data['bas_0.0'][row]}
        for conc in ['250.0','125.0','50.0','25.0','10.0','5.0','2.5']:
            median_vals['occlusion'][loc][conc]=cAMP_loaded_data['ind_'+conc][row]
```

```python
plt.figure(figsize=(6,6))
    
labels={'250.0':'250uM', '125.0':'125uM', '50.0':'50uM', '25.0':'25uM', '10.0':'10uM', '5.0':'5uM', '2.5':'2.5uM'}
colors={'250.0':'midnightblue', '125.0':'navy', '50.0':'mediumblue', '25.0':'b', '10.0':'royalblue', '5.0':'cornflowerblue', '2.5':'lightskyblue'}
lib='c61r18'
out_lab='outliers'
    
for construct in median_vals[lib]: 
    if np.isnan(median_vals[lib][construct]['0.0']): continue
    if construct in outliers: 
        plt.scatter(median_vals[lib][construct]['0.0'],median_vals[lib][construct]['250.0'],label=out_lab, color='gray')
        out_lab=''
        continue
    
    for conc in median_vals[lib][construct]:
        if conc=='0.0': continue
        if np.isnan(median_vals[lib][construct][conc])==False:
            plt.scatter(median_vals[lib][construct]['0.0'], median_vals[lib][construct][conc], label=labels[conc], color=colors[conc], alpha=0.6)
            labels[conc]=''
            

plt.yscale('log')
plt.xscale('log')
plt.xlabel('activity without CRP (au)')
plt.ylabel('activity with CRP (au)')
plt.title('CRP site at -61.5\n')
x_min, x_max = 0.001, 100
plt.xlim(x_min,x_max)
plt.ylim(x_min,x_max)
plt.plot([x_min, x_max], [x_min, x_max], linestyle='--', color='k', zorder=1)
plt.legend(loc='lower right', scatterpoints=1, prop={'size':10})
plt.show()
```

```python
c61_tsat=cAMP_summary['tsat'][0]
c61_tbg=cAMP_summary['tbg'][0]
c61_alpha=cAMP_summary['alpha'][0]

c71_tbg=cAMP_summary['tbg'][1]
c71_alpha=cAMP_summary['alpha'][1]

oc_tbg=cAMP_summary['tbg'][2]
oc_alpha=cAMP_summary['alpha'][2]

knowns={'c61r18':[c61_tsat,c61_tbg,c61_alpha],
        'c71r18':[c61_tsat,c71_tbg,c71_alpha],
        'occlusion':[c61_tsat,oc_tbg,oc_alpha]
       }

F_vals={'c61r18':{},'c71r18':{},'occlusion':{}}
for conc in ['250.0', '125.0','50.0','25.0', '10.0', '5.0', '2.5']:
    F_vals['c61r18'][conc]=cAMP_summary['F_'+conc][0]
    F_vals['c71r18'][conc]=cAMP_summary['F_'+conc][1]
    F_vals['occlusion'][conc]=cAMP_summary['F_'+conc][2]

fitrs = np.linspace(-200,5,1000)

fitxs_fp={}
fitys_fp={}

for constr in ['c61r18','c71r18','occlusion']:
    fitxs_fp[constr]={}
    fitys_fp[constr]={}
    
    tsat=knowns[constr][0]
    tbg=knowns[constr][1]
    alpha=knowns[constr][2]
    
    for conc in ['250.0', '125.0','50.0','25.0', '10.0', '5.0', '2.5']:
        F=F_vals[constr][conc]
        
        fitxs_fp[constr][conc]=[(tsat*(np.exp(s)/(1+np.exp(s)))+tbg) for s in fitrs]
        fitys_fp[constr][conc]=[tsat*((np.exp(s)+alpha*F*np.exp(s))/(1+np.exp(s)+F+alpha*F*np.exp(s)))+tbg for s in fitrs]
```

```python
plt.figure(figsize=(6,6))
    
labels={'250.0':'250uM', '125.0':'125uM', '50.0':'50uM', '25.0':'25uM', '10.0':'10uM', '5.0':'5uM', '2.5':'2.5uM'}
colors={'250.0':'midnightblue', '125.0':'navy', '50.0':'mediumblue', '25.0':'b', '10.0':'royalblue', '5.0':'cornflowerblue', '2.5':'lightskyblue'}
lib='c61r18'
out_lab='outliers'
    
for construct in median_vals[lib]: 
    if np.isnan(median_vals[lib][construct]['0.0']): continue
    if construct in outliers: 
        plt.scatter(median_vals[lib][construct]['0.0'],median_vals[lib][construct]['250.0'],label=out_lab, color='gray')
        out_lab=''
        continue
    
    for conc in median_vals[lib][construct]:
        if conc=='0.0': continue
        if np.isnan(median_vals[lib][construct][conc])==False:
            plt.scatter(median_vals[lib][construct]['0.0'], median_vals[lib][construct][conc], label=labels[conc], color=colors[conc], alpha=0.6)
            labels[conc]=''

        
#for construct in library_groups[lib]['wt']: 
#    if construct not in median_vals: continue
#    plt.scatter(median_vals[construct]['0.0'], median_vals[construct]['250.0'], color='black', alpha=0.7)

for conc in fitxs_fp[lib]:
    plt.plot((fitxs_fp[lib][conc]),(fitys_fp[lib][conc]), color=colors[conc])

plt.yscale('log')
plt.xscale('log')
plt.xlabel('activity without CRP (au)')
plt.ylabel('activity with CRP (au)')
plt.title('CRP site at -61.5\n fitting F and P')
x_min, x_max = 0.001, 100
plt.xlim(x_min,x_max)
plt.ylim(x_min,x_max)
plt.plot([x_min, x_max], [x_min, x_max], linestyle='--', color='k', zorder=1)
plt.legend(loc='lower right', scatterpoints=1, prop={'size':10})
plt.show()
```
